```python
import datetime
import csv

# 日時の取得
now = datetime.datetime.now()
# ディレクトリの指定はここ
filename = './' + "trainingdata" + '.csv'

# ファイル，1行目(カラム)の作成
with open(filename, 'w') as f:
    writer = csv.writer(f)
    writer.writerow(['filenames','abnormal'])

for i in range(1,10):
    with open(filename, 'a', newline="") as f:
        writer = csv.writer(f)
        fname = 'awake_0000' + str(i) + '.jpg'
        writer.writerow([fname,0])
        
for i in range(10,100):
    with open(filename, 'a', newline="") as f:
        writer = csv.writer(f)
        fname = 'awake_000' + str(i) + '.jpg'
        writer.writerow([fname,0])
        
for i in range(100,200):
    with open(filename, 'a', newline="") as f:
        writer = csv.writer(f)
        fname = 'awake_00' + str(i) + '.jpg'
        writer.writerow([fname,0])
    # 処理が終わったらbreak

for i in range(1,10):
    with open(filename, 'a', newline="") as f:
        writer = csv.writer(f)
        fname = 'sleep_0000' + str(i) + '.jpg'
        writer.writerow([fname,1])
        
for i in range(10,100):
    with open(filename, 'a', newline="") as f:
        writer = csv.writer(f)
        fname = 'sleep_000' + str(i) + '.jpg'
        writer.writerow([fname,1])
        
for i in range(100,200):
    with open(filename, 'a', newline="") as f:
        writer = csv.writer(f)
        fname = 'sleep_00' + str(i) + '.jpg'
        writer.writerow([fname,1])
    # 処理が終わったらbreak
```
